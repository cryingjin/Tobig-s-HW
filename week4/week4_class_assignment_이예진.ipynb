```python
import random
class MyDataLoader:
    def __init__(self,data):
        self.size = len(data)
        self.data = data
        
    def __iter__(self):
        self.index = 0
        return self
    
    def __next__(self):
        if self.index>=self.size:
            raise StopIteration
        n = self.data[self.index]
        self.index += 1
        return n
    
    def shuffle(self):
        random.shuffle(self.data)
```

```python
loader = MyDataLoader([i for i in range(300)])
loader.shuffle()
for i, x in enumerate(loader):
    print("{}:{}".format(i,x))
```
